# **Loading the Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/happiness_2017 .csv')

# Inspect the first few rows and column information
print(df.head())
print(df.info())
```

# **Descriptive Statistics**

```python
stats = df.describe()
stats.to_csv('descriptive_statistics.csv')
print(stats)
```

# **Lineplot**

```python
plot_df = df.dropna(subset=['HappinessScore', 'Region', 'Log GDP per capita'])
plt.figure(figsize=(10, 6))
sns.lineplot(data=plot_df, x='Rank', y='HappinessScore')
plt.title('Lineplot: Happiness Score vs Rank')
plt.show()
```

# **Barplot**

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=plot_df, x='Region', y='HappinessScore', estimator='mean', errorbar=None)
plt.xticks(rotation=45)
plt.title('Barplot: Mean Happiness Score per Region')
plt.show()
```

# **Heatmap**

```python
plt.figure(figsize=(12, 10))
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap: Correlation Matrix of Variables')
plt.show()
```

# **Scatter Plot**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=plot_df, x='Log GDP per capita', y='HappinessScore', hue='Region')
plt.title('Scatterplot: Log GDP per capita vs Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# **Histogram Plot**

```python
plt.figure(figsize=(10, 6))
sns.histplot(plot_df['HappinessScore'], kde=True, color='skyblue')
plt.title('Histplot: Distribution of Happiness Score')
plt.show()
```

# **Swarm Plot**

```python
plt.figure(figsize=(14, 8))
sns.swarmplot(data=plot_df, x='Region', y='HappinessScore')
plt.xticks(rotation=45)
plt.title('Swarmplot: Happiness Score across Regions')
plt.show()
```

# **Box plot**

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=plot_df, x='Region', y='HappinessScore')
plt.xticks(rotation=45)
plt.title('Boxplot: Happiness Score across Regions')
```
